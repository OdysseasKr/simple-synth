# Generating Sounds
In this example we will see how to generate a couple of simple sounds using the synthesizer.

```python
from simplesynth.synth import Synth
import matplotlib.pyplot as plt
import sounddevice as sd
import time

# Initialise the synth object
s = Synth()
```

First, let's make a simple sine wave

```python
s.set_parameters(osc_1='Sine', mix=0) # Setting the mix to 0 makes the synth ignore the second osc
# Receive a numpy array with the sound of frequency 440Hz and duration 1sec 
arr = s.get_sound_array(note=440, duration=1)
plt.plot(arr[:500])
plt.show()
```

You can also listen to the sound using _sounddevice_.

```python
sd.play(arr, s.sf)
time.sleep(1)
```

Next, let's add a second oscillator and some filtering

```python
s.set_parameters(osc_1='Sine', mix=0.7, osc_2='Square')
arr = s.get_sound_array(note=440, duration=1)
plt.plot(arr[:500])
plt.show()
```

```python
sd.play(arr, s.sf)
time.sleep(1)
```
